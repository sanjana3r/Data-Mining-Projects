```python
# Imported for data management (dataframes)
import pandas as pd

# Imported to create word clouds
# This package needs part of anaconda and needs to be installed
#    conda install -c conda-forge wordcloud
from wordcloud import WordCloud

# Imported to allow for the display of word clouds
import matplotlib.pyplot as plt
```

## Concept Description:
Determine the difference in the language used to discuss Data Mining and the language used to discuss Artificial Intelligence. 

## Data Collection: 
Google the search phrase “What is Data Mining” and “What is Artificial Intelligence”.  For each search phrase, the first 15 results pages are browsed and search for a block of text that starts with something akin to “Data Mining is …” or “Artificial Intelligence is …”.  The text blocks are recorded as examples along with the link of the page, the search phrase, and the date the results were pulled, and the name of the site the results were from.

The results for each search phrase are saved in one comma separated file per each search phrase.

## Example Description:
<b>Category</b>  
The search phrase that is searched.  The Category is a label.

<b>PullDate</b>  
The date the search result was obtained.  The pull date is a date.

<b>Source</b>  
The name of the site that was returned as a result. The source is a character string.

<b>Link</b>  
The link that was returned as a result.  The link is a character string.

<b>Text</b>  
The text block that is was selected from the page that was returned as a result of the search.  The text is a character string.

## Data Import and Wrangling: 

The results of each search is read from the respective comma separated value file (csv) into separate dataframes.  Careful attention is paid to make sure the data is read in as character strings.

```python
df_mining = pd.read_csv("Data_Mining.csv", dtype=str,engine ='python')
df_ai = pd.read_csv("Artificial-Intelligence.csv", dtype=str, engine ='python')
df_mining
```

```python
df_ai
```

The text blocks from the results of a search is concatenated together into one large text string.  

The blank text blocks were loaded in as "Na".  We convert them back to empty text strings because WordCloud will not accept non-string data.

```python
# Convert all Text attribute values to text strings.
str_mining = ''.join(list(df_mining['Text']))

# Replace any Na values with a blank text string.
str_ai = ''.join(list(df_ai['Text'].fillna('')))
```

## Exploratory Data Analysis:
#### 1) I tried counting the frequency of the of the words of "Text" Attribute of both Data Mining and Artificial Intelligence.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(analyzer='word')
data=cv.fit_transform(df_mining['Text'])
df_dtm = pd.DataFrame(data.toarray(), columns=cv.get_feature_names())
df_dtm.index=df_mining['Text']
df_dtm.head(3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(analyzer='word')
data=cv.fit_transform(df_ai['Text'])
df_dtm = pd.DataFrame(data.toarray(), columns=cv.get_feature_names())
df_dtm.index=df_ai['Text']
df_dtm.head(3)
```

#### 2) Using TextBlob to Review highest polarity for both Data Mining and Artificial Intelligence

```python
from textblob import TextBlob
df_mining['polarity']=df_mining['Text'].apply(lambda x:TextBlob(x).sentiment.polarity)
df_ai['polarity']=df_ai['Text'].apply(lambda x:TextBlob(x).sentiment.polarity)
```

```python
print("3 Random Reviews with Highest Polarity for Data Mining:\n")
for index,review in enumerate(df_mining.iloc[df_mining['polarity'].sort_values(ascending=False)[:3].index]['Text']):
    print('Review {}:\n'.format(index+1),review)
print("\n--------------------------------------------------------------------------------------------------------\n")
print("3 Random Reviews with Highest Polarity for Artificial Intelligence:\n")
for index,review in enumerate(df_ai.iloc[df_ai['polarity'].sort_values(ascending=False)[:3].index]['Text']):
    print('Review {}:\n'.format(index+1),review)
```

#### 3) Plotting the results of polarity in a bar graph

```python
import numpy as np
product_polarity_sorted=pd.DataFrame(df_mining.groupby('Source')['polarity'].mean().sort_values(ascending=True))

plt.figure(figsize=(16,8))
plt.xlabel('Polarity')
plt.ylabel('Source')
plt.title('Polarity of Text in Data Mining')
polarity_graph=plt.barh(np.arange(len(product_polarity_sorted.index)),product_polarity_sorted['polarity'],color='purple',)
# Writing product names on bar
for bar,product in zip(polarity_graph,product_polarity_sorted.index):
  plt.text(0.005,bar.get_y()+bar.get_width(),'{}'.format(product),va='center',fontsize=11,color='white')

# Writing polarity values on graph
for bar,polarity in zip(polarity_graph,product_polarity_sorted['polarity']):
  plt.text(bar.get_width()+0.001,bar.get_y()+bar.get_width(),'%.3f'%polarity,va='center',fontsize=11,color='black')
  
plt.yticks([])
plt.show()
```

```python
import numpy as np
product_polarity_sorted=pd.DataFrame(df_ai.groupby('Source')['polarity'].mean().sort_values(ascending=True))

plt.figure(figsize=(16,8))
plt.xlabel('Polarity')
plt.ylabel('Source')
plt.title('Polarity of Text in Data Mining')
polarity_graph=plt.barh(np.arange(len(product_polarity_sorted.index)),product_polarity_sorted['polarity'],color='purple',)
# Writing product names on bar
for bar,product in zip(polarity_graph,product_polarity_sorted.index):
  plt.text(0.005,bar.get_y()+bar.get_width(),'{}'.format(product),va='center',fontsize=11,color='white')

# Writing polarity values on graph
for bar,polarity in zip(polarity_graph,product_polarity_sorted['polarity']):
  plt.text(bar.get_width()+0.001,bar.get_y()+bar.get_width(),'%.3f'%polarity,va='center',fontsize=11,color='black')
  
plt.yticks([])
plt.show()
```
